```python
## g/pot

import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import plotly.graph_objects as go

# 数据集，这里修正了数据变量名的笔误
data = {
    'mass_g': [1.16, 0.84, 1.28, 1.17, 0.80, 0.97, 1.26, 1.33, 1.24, 1.74, 1.68, 1.77, 3.94, 3.13, 3.32, 4, 4.7, 5, 5.7, 6.2, 6.7, 7.2, 7.23, 7.23, 7.25, 7.25, 7.26, 7.27, 7.27, 7.26, 7.27 ,7.26, 7.27, 7.27, 7.26, 7.27],
    'OLIV_g': [0, 0, 0, 0, 0, 0, 10, 10, 10, 10, 10, 10, 20, 20, 20, 25, 25, 30, 35, 35, 40, 40, 40, 45, 47, 50, 50, 55, 60, 70, 80, 90, 100, 100, 100, 100],
    'KNO3_g': [0, 0, 0, 1, 1, 1, 0, 0, 0, 1, 1, 1, 1, 1, 1, 2, 3, 3.5, 4, 4.3, 4.4 ,4.54, 4.7, 4.8, 5.0, 6, 6, 6.5, 6.5 ,7, 7 ,7, 7.5, 7.5, 8, 8]
}
df = pd.DataFrame(data)

# 定义自变量和因变量
X = df[['OLIV_g', 'KNO3_g']]
y = df['mass_g']

# 使用多项式特征来模拟非线性关系，并进行线性回归
degree = 2  # 使用2次多项式
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
polyreg.fit(X, y)



# 提取系数和截距
coefficients = polyreg.named_steps['linearregression'].coef_
intercept = polyreg.named_steps['linearregression'].intercept_

# 构建方程式字符串
equation = f"mass_g = {intercept:.4f}"
for i, coef in enumerate(coefficients[1:], start=1):
    equation += f" + ({coef:.4f})*x{i}"


# 准备3D图形的数据
x_range = np.linspace(df['OLIV_g'].min(), df['OLIV_g'].max(), 30)
y_range = np.linspace(df['KNO3_g'].min(), df['KNO3_g'].max(), 30)
xx, yy = np.meshgrid(x_range, y_range)
zz = polyreg.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
zz[zz < 0] = 0  # 确保不会有负值

# 使用Plotly创建交互式3D图，添加等高线
fig = go.Figure(data=[go.Surface(
    z=zz, x=xx, y=yy, colorscale='Viridis',
    contours={"z": {"show": True, "start": 1.0, "end": 8.0, "size": 0.5, "color":"white"}}
)])
fig.add_trace(go.Scatter3d(x=df['OLIV_g'], y=df['KNO3_g'], z=df['mass_g'], mode='markers', marker=dict(size=5, color='red')))
fig.add_annotation(text=equation, xref="paper", yref="paper", x=0.5, y=0.95, showarrow=False, font=dict(size=12))

# 更新图形的布局
fig.update_layout(
    title=dict(
        text='空心菜之生長曲面圖',
        x=0.5,  # Center the title
        y=0.95,  # Adjust the vertical position if needed
        xanchor='center',  # Anchor the title at the center
        yanchor='top',  # Anchor the title at the top
        font=dict(size=24)  # Increase the font size
    ),
    scene=dict(
        xaxis_title='橄欖石 (克)',
        yaxis_title='硝酸鉀 (克)',
        zaxis=dict(title='重量 (克)', range=[0, zz.max()])  # 设置Z轴的下限为0
    )

)

# 更新注释的字体大小
# fig.update_annotations(font=dict(size=18))  # Increase the font size of the equation


# 显示图形
fig.show()
print(coefficients)
print(equation)
```
